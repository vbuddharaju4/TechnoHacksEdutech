**TASK 1 : Perform Data Cleaning**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data1 = pd.read_csv('/content/train.csv')
data2 = pd.read_csv('/content/test.csv')
data3 = pd.read_csv('/content/gender_submission.csv')
```

```python
data_cleaned1 = data1.dropna()
```

```python
data_cleaned2 = data2.dropna()
```

```python
data_cleaned3 = data3.dropna()
```

```python
plt.figure(figsize=(8, 6))
sns.scatterplot(data_cleaned1)['Age']
plt.xlabel('Age')
plt.title('Scatterplot Of Age')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.scatterplot(data_cleaned2)['Age']
plt.xlabel('Age')
plt.title('Scatterplot Of Age')
plt.show()

plt.figure(figsize=(8, 6))
sns.scatterplot(data_cleaned2)['Fare']
plt.xlabel('Fare')
plt.title('Scatterplot Of Fare')
plt.show()
```

```python
age_iqr = data_cleaned1['Age'].quantile(0.75) - data_cleaned1['Age'].quantile(0.25)
age_lower = data_cleaned1['Age'].quantile(0.25) - 1.5 * age_iqr
age_upper = data_cleaned1['Age'].quantile(0.75) + 1.5 * age_iqr
data_cleaned1 = data_cleaned1[(data_cleaned1['Age'] >= age_lower) & (data_cleaned1['Age'] <= age_upper)]
```

```python
age_iqr = data_cleaned2['Age'].quantile(0.75) - data_cleaned2['Age'].quantile(0.25)
age_lower = data_cleaned2['Age'].quantile(0.25) - 1.5 * age_iqr
age_upper = data_cleaned2['Age'].quantile(0.75) + 1.5 * age_iqr
data_cleaned2 = data_cleaned2[(data_cleaned2['Age'] >= age_lower) & (data_cleaned2['Age'] <= age_upper)]
```

```python
age_iqr = data_cleaned2['Fare'].quantile(0.75) - data_cleaned2['Fare'].quantile(0.25)
age_lower = data_cleaned2['Fare'].quantile(0.25) - 1.5 * age_iqr
age_upper = data_cleaned2['Fare'].quantile(0.75) + 1.5 * age_iqr
data_cleaned2 = data_cleaned2[(data_cleaned2['Fare'] >= age_lower) & (data_cleaned2['Fare'] <= age_upper)]
```

 **TASK 2 : Calculate Summary Statistics**

```python
mean_values = data_cleaned1.mean()
median_values = data_cleaned1.median()
mode_values = data_cleaned1.mode().iloc[0]
std_values = data_cleaned1.std()
print("Summary Statistics:")
print("Mean:")
print(mean_values)
print("\nMedian:")
print(median_values)
print("\nMode:")
print(mode_values)
print("\nStandard Deviation")
print(std_values)
```

```python
mean_values = data_cleaned2.mean()
median_values = data_cleaned2.median()
mode_values = data_cleaned2.mode().iloc[0]
std_values = data_cleaned2.std()
print("Summary Statistics:")
print("Mean:")
print(mean_values)
print("\nMedian:")
print(median_values)
print("\nMode:")
print(mode_values)
print("\nStandard Deviation")
print(std_values)
```
